# **7. SELEÇÃO DE ATRIBUTOS**

Por estarmos tratando de um problema de clasificação, iremos usar a técnica de florestas (conjunto de árvores de decisão) que foi dado como exemplo no material extra disponibilizado no GitHub da disciplina.

```
# Importação das bibliotecas para fazer a manipulação nos dados e seleção

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier

from google.colab import drive 
drive.mount('/content/gdrive/')
```

```
# Importação do dataset pré-processado
df = pd.read_csv("/content/gdrive/MyDrive/Colab Notebooks/all_seasons_pre_processing.csv")
```

```
# Visuzalizando os dados para checar se deu certo a importação
df.head()
```

```
# Reorganizando tabelas e definindo target para poder usar no seletor
#discard = [ 'player_id', 'gp', 'pts', 'reb', 'ast', 'net_rating', 'oreb_pct', 'dreb_pct', 'usg_pct', 'ts_pct', 'ast_pct', 'season']

X = df.drop('was_mvp', axis=1)
#y = df.drop(discard, axis=1)
y = pd.Series(df.was_mvp, index = X.index, name = "was_mvp")
X.head()
```

```
# Checando se a coluna was_mvp virou um Series
y.head()
```

```
# Declarando o seletor de florestas
oneshot_selector = SelectFromModel(RandomForestClassifier(),threshold="median", max_features=6)
# Criando novo DF com os dados selecionados
n_df = pd.DataFrame(oneshot_selector.fit_transform(X, y), 
             columns=X.columns[oneshot_selector.get_support()])
n_df.head()
```

```
n_df.duplicated().sum()
```

```
# Juntando was_mvp ao novo DF
n_df = n_df.join(y)
n_df.head()
```

```
# Verificando duplicatas
n_df.duplicated().sum()
```

```
# Escrevendo os dados selecionados em .csv
n_df.to_csv("/content/gdrive/MyDrive/Colab Notebooks/BaseReduzida2.csv", index=False)
```
